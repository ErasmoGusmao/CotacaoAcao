#### Pegando a cotação de 1 ação

```python
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt

cotacao_bovespa = web.DataReader('^BVSP', data_source ='yahoo', start='01/01/2020', end = '01/01/2021') # formato da data mês/dia/ano ou mês-dia-ano
display(cotacao_bovespa)
cotacao_bovespa['Adj Close'].plot(figsize=(15,10))
plt.show()

cotacao_BTC = web.DataReader('BTC-USD', data_source ='yahoo', start='01/01/2020', end = '01/01/2021') # formato da data mês/dia/ano ou mês-dia-ano
display(cotacao_BTC)
cotacao_BTC['Adj Close'].plot(figsize=(15,10))
plt.show()
```

#### Pegando a cotação de uma carteira de ações em em intervalo de datas específica

```python
tabela_empreas = pd.read_excel('Empresas.xlsx')
display(tabela_empreas)

for empresa in tabela_empreas['Empresas']:
    print(f'============================================ {empresa} ======================================================')
    cotacao = web.DataReader(f'{empresa}.SA', data_source ='yahoo', start='01/01/2020', end = '01/01/2021') # formato da data mês/dia/ano ou mês-dia-ano
    display(cotacao)
    cotacao['Adj Close'].plot(figsize=(15,10))
    plt.show()
```
